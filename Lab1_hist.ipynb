```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the image in RGB format
image2 = cv2.imread('im2.png')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Create an empty matrix of dimensions 1x256
histogram = np.zeros((1, 256))

# Calculate the histogram
height, width = gray_image.shape
for i in range(height):
    for j in range(width):
        pixel_value = gray_image[i, j]
        histogram[0, int(pixel_value)] += 1

# Display the histogram
plt.bar(range(256), histogram[0])
plt.xlabel('Gray Level')
plt.ylabel('Frequency')
plt.title('Grayscale Image Histogram')
plt.show()
```

```python
# Calculate the total number of pixels
total_pixels = gray_image.shape[0] * gray_image.shape[1]

# Calculate pr(rk) for each gray level
histogram = np.histogram(gray_image.flatten(), bins=256, range=(0, 256))[0]
pr = histogram / total_pixels

# Print pr(rk) for each gray level
for rk, pr_rk in enumerate(pr):
    print(f"pr({rk}): {pr_rk}")
```

```python
# Calculate the cumulative distribution function (CDF)
cdf = np.cumsum(pr)

# Print the CDF values
for rk, cdf_rk in enumerate(cdf):
    print(f"CDF({rk}): {cdf_rk}")
```

```python
# Normalize the CDF to range [0, 255]
cdf_normalized = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())

# Map the original gray levels to the new gray levels
equalized_histogram = np.round(cdf_normalized).astype(int)

# Print the equalized histogram mapping
for rk, sk in enumerate(equalized_histogram):
    print(f"Original Gray Level {rk} -> New Gray Level {sk}")
```

```python
# Apply histogram equalization
equalized_image = equalized_histogram[gray_image]

# Display the original and equalized images side by side
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(gray_image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')
axs[1].imshow(equalized_image, cmap='gray')
axs[1].set_title('Equalized Image')
axs[1].axis('off')
plt.tight_layout()
plt.show()

```

```python
# Plot the histogram for the equalized image
plt.bar(range(256), equalized_histogram)
plt.xlabel('Gray Level')
plt.ylabel('Frequency')
plt.title('Equalized Histogram')
plt.show()
```
